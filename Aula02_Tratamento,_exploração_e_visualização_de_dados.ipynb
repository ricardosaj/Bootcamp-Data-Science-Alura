# Desafio: Criar uma coluna extra mostrando a região de cada estado

```python
#regiao = usaveis.index.str.slice(stop=1).astype(int) 
#usaveis.insert(0, "Nome da Região", regiao)
traducaoReg = []
for i in usaveis.index:
  if i[0]=='1':
    traducaoReg.append('Norte')
  elif i[0]=='2':
    traducaoReg.append('Nordeste')
  elif i[0]=='3':
    traducaoReg.append('Sudeste')
  elif i[0]=='4':
    traducaoReg.append('Sul')
  elif i[0]=='5':
    traducaoReg.append('Centro-Oeste')
  else:
    traducaoReg.append('Erro')
  
usaveis['Nome da Região'] = traducaoReg
```

```python
#usaveis = usaveis.drop('Regiao', axis=1) 
usaveis
```

```python
totalGastoRegioes = usaveis.sum(axis=1)
usaveis.insert(0, "Total", totalGastoRegioes)
usaveis
```
